```python
import pandas as pd
import numpy as np

def clean_locality_column(df, column_name='locality'):
    """Clean locality column by extracting unique city names and removing duplicates"""
    
    # Create a copy to avoid modifying original
    df_clean = df.copy()
    
    # Step 1: Extract just the city name (first part before first comma)
    df_clean['locality_clean'] = df_clean[column_name].str.split(',').str[0].str.strip().str.lower()
    
    # Step 2: Handle common city variations and normalization
    locality_mapping = {
        'new york city': 'new york',
        'nyc': 'new york',
        'la': 'los angeles',
        'san fran': 'san francisco',
        'sf': 'san francisco',
        'dc': 'washington dc',
        'washingtondc': 'washington dc',
        'washingtond.c.': 'washington dc',
        'bombay': 'mumbai',  # Based on your data
        'greater london': 'london'
    }
    
    df_clean['locality_clean'] = df_clean['locality_clean'].replace(locality_mapping)

    return df_clean

# Clean the locality column
df_clean = clean_locality_column(df)

print("Original data shape:", df.shape)
print("Cleaned data shape:", df_clean.shape)
print("\nCleaned locality examples:")
print(df_clean[['locality', 'locality_clean']].head(10))
```

```python
df = df_clean.drop(columns=['locality'])
```

```python
# Convert float to integer year
df['year founded'] = df['year founded'].astype(int)
print("Year founded as integers:")
print(df[['name', 'year founded']].head())
```

```python
df.head()
```

```python
#Standardizing the column names
df.columns = (df.columns.str.strip().str.lower().str.replace(' ', '_'))
```

```python
df.rename(columns={'unnamed:_0': 'company_id'}, inplace=True)
```

```python
#Trying to detect and remove outlier which are impossible values in this case
df = df[(df['year_founded'] >= 1000) & (df['year_founded'] <= 2025)]
df = df[(df['current_employee_estimate'] > 0) & (df['total_employee_estimate'] > 0)]
```

```python
#A metic used to calculate the age of a company
from datetime import datetime
df['company_age'] = datetime.now().year - df['year_founded']
```

```python
#A company growth indicator which indicates how much a company has grown in ratio and rounded to three decimal points.
df['employee_growth_ratio'] = ((df['total_employee_estimate'] - df['current_employee_estimate'])/ df['total_employee_estimate']).round(3)
```

```python
# Save the cleaned dataframe to a new CSV
df.to_csv("companies_sorted_cleaned.csv", index=False)
```

```python
import pandas as pd
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv

# Load the secret file
load_dotenv()

# Database connection info from .env file
user = os.getenv('DB_USER')
password = os.getenv('DB_PASSWORD')
host = os.getenv('DB_HOST')
port = os.getenv('DB_PORT')
database = os.getenv('DB_NAME')

# Create SQLAlchemy engine
engine = create_engine(f'postgresql://{user}:{password}@{host}:{port}/{database}')

# CSV file name
file_path = 'companies_sorted_cleaned.csv'

# Table name in Postgres
table_name = 'companie_data'

# Read and import in chunks
chunksize = 200_000  # adjust based on your RAM

print(" Starting import...")
for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunksize)):
    chunk.to_sql(table_name, engine, index=False, if_exists='append')
    print(f" Imported chunk {i+1} ({len(chunk)} rows)")

print(" All data imported successfully into 'companie_data' table!")
```
